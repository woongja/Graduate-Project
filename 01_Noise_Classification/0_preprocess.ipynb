```python
import glob
import os
import pandas as pd

METADATA_DIR = "/home/woongjae/ADD_LAB/Graduate-Project/00_Dataset_gen/Datasets/noise_dataset/NC"
OUT_DIR      = "/home/woongjae/ADD_LAB/Graduate-Project/01_Noise_Classification/protocols"
os.makedirs(OUT_DIR, exist_ok=True)

def normalize_label(aug_type):
    if aug_type in ('high_pass_filter', 'low_pass_filter'):
        return 'band_pass_filter'
    return aug_type

def write_protocol(df, path):
    with open(path, 'w') as f:
        for row in df.itertuples(index=False):
            f.write(f"{row.file_path} {row.split} {row.label}\n")

# ── Load all CSVs ──────────────────────────────────────────────
all_dfs = []
for csv_path in sorted(glob.glob(f"{METADATA_DIR}/metadata_*.csv")):
    df = pd.read_csv(csv_path, dtype=str)
    all_dfs.append(df[['file_path', 'aug_type', 'split']])

df_all = pd.concat(all_dfs, ignore_index=True)
df_all['label'] = df_all['aug_type'].map(normalize_label)

# ── train_protocol.txt  (train + dev) ─────────────────────────
train_df = df_all[df_all['split'].isin(['train', 'dev'])][['file_path', 'split', 'label']]
write_protocol(train_df, f"{OUT_DIR}/train_protocol.txt")
print(f"train_protocol.txt  rows={len(train_df)}")
print(train_df['split'].value_counts().to_string())
print(train_df['label'].value_counts().to_string())

# ── eval_protocol.txt  (eval) ──────────────────────────────────
eval_df = df_all[df_all['split'] == 'eval'][['file_path', 'split', 'label']]
write_protocol(eval_df, f"{OUT_DIR}/eval_protocol.txt")
print(f"\neval_protocol.txt   rows={len(eval_df)}")
print(eval_df['label'].value_counts().to_string())
```
