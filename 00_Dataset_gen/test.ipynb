```python
import librosa, time

utt_id = "/home/woongjae/noise-tracing/multi-feature_fusion/Datasets/datasets/spoof/noise/background_noise/p228_to_p227_75_background_noise.wav"
t0 = time.time()
try:
    wav, fs = librosa.load(utt_id, sr=16000)
    print("librosa.load:", wav.shape, time.time() - t0, "sec")
    t1 = time.time()
    # stft
    stft = librosa.stft(wav, n_fft=2048, hop_length=512, window='hamming')
    print("stft:", stft.shape, time.time() - t1, "sec")
    # mfcc
    frame_size = int(0.025 * 16000)
    hop_size = int(0.010 * 16000)
    mfcc = librosa.feature.mfcc(y=wav, sr=16000, n_mfcc=13, n_fft=frame_size, hop_length=hop_size, n_mels=128)
    print("mfcc:", mfcc.shape, time.time() - t1, "sec")
    # f0
    t2 = time.time()
    f0 = librosa.yin(wav, fmin=50, fmax=600, sr=16000, frame_length=2048, hop_length=512)
    print("yin:", f0.shape, time.time() - t2, "sec")
except Exception as e:
    print("Error:", e)
```
